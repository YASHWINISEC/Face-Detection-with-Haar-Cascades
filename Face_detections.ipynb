```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rect=face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,0),10)
        cv2.imshow("Face Detection", face_img)
    return face_img
```

```python
result = detect_face(img1)
```

```python
plt.imshow(result, cmap="gray")
plt.show()
```

```python
result1 = detect_face(img2)
plt.imshow(result1,cmap="gray")
plt.show()
```

```python
result2 = detect_face(img3)
plt.imshow(result2,cmap="gray")
plt.show()
```

```python
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')

def detect_eye(im):
    face=im.copy()
    rect=eye_cascade.detectMultiScale(face)

    for (x,y,w,h) in rect:
        cv2.rectangle(face,(x,y) , (x+w,y+h) , (127,0,255) , 10)
        cv2.imshow('Face Detection',face)
    return face
```

```python
result= detect_eye(img1) 
plt.imshow(result,cmap='gray')
plt.show()
```

```python
result= detect_eye(img2) 
plt.imshow(result,cmap='gray')
plt.show()
```

```python
result= detect_eye(img3) 
plt.imshow(result,cmap='gray')
plt.show()
```

```python
cap = cv2.VideoCapture(0)
plt.ion()
fig, ax = plt.subplots()

ret, frame = cap.read(0)
frane = detect_face(frame)
im=ax.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title('Video Face Detection')

while True:
    ret, frame = cap.read(0)
    frame = detect_face(frame)
    im.set_data(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.pause(0.10)

cap.release()
plt.close()
```
